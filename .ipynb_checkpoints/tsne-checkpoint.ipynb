```python
from PIL import Image
import os
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from tensorflow import keras
from sklearn.manifold import TSNE
```

```python
model = keras.applications.vgg16.VGG16(include_top=False, weights='imagenet', pooling = 'avg', input_tensor=None, input_shape=(128, 128, 3))
```

```python
model.summary()
```

```python
model.compile(optimizer = 'adam', loss = 'mae')
```

```python
# Open images
IMG_SIZE = (128, 128)
NUM_SAMPLES = 1000
BASE_IMG_PATH = './imageExtractor'
samples = []
for i, filename in enumerate(os.listdir(BASE_IMG_PATH)):
    if filename.strip()[-4:] == '.jpg':
        img = Image.open(os.path.join(BASE_IMG_PATH, filename))
        img = img.resize(IMG_SIZE)
        samples.append(np.array(img))
        
    if i > NUM_SAMPLES:
        break
```

```python
_, axes = plt.subplots(10, 10, figsize = (20, 20))
axes = axes.flatten()
for i in range(len(axes)):
    axes[i].imshow(samples[i])
```

```python
features = model.predict(np.array(samples, dtype = 'float'), verbose = 1)
```

```python
features.shape
```

```python
tsne = TSNE()
embeddings = tsne.fit_transform(features)
```

```python
embeddings.shape
```

```python
plt.figure(figsize = (10, 10))
plt.scatter(embeddings[:, 0], embeddings[:, 1], s = 1)
plt.show()
```
